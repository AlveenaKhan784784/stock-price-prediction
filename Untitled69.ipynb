# Stock Price Prediction using Machine Learning

## Objective
The objective of this project is to predict future stock prices based on historical data using machine learning algorithms. Specifically, we will:
1. Analyze and preprocess historical stock data.
2. Detect and remove outliers to clean the dataset.
3. Engineer features for the predictive model.
4. Build and evaluate models using Linear Regression and Random Forest Regressor.
5. Compare model performance and visualize results.

## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

## Data Preprocessing

```python
# Convert the date column to datetime
df['date'] = pd.to_datetime(df['date'])

# Keep only relevant columns
df = df[['date', 'close', 'open', 'high', 'low', 'volume', 'Name']]

# Filter for one stock (Apple 'AAPL')
apple_df = df[df['Name'] == 'AAPL']

# Plot the stock prices over time
plt.figure(figsize=(10,6))
plt.plot(apple_df['date'], apple_df['close'], label='Close Price')
plt.title('Apple Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

## Outlier Detection and Removal

```python
# Calculate the IQR for 'close' prices
Q1 = apple_df['close'].quantile(0.25)
Q3 = apple_df['close'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = apple_df[(apple_df['close'] < lower_bound) | (apple_df['close'] > upper_bound)]

# Display outliers
print(f'Number of outliers: {len(outliers)}')
print(outliers.head())
```

```python
# If it has outliers then we Remove them like:
apple_df_clean = apple_df[(apple_df['close'] >= lower_bound) & (apple_df['close'] <= upper_bound)]

# Check the shape of the cleaned dataset
print(f'Original dataset shape: {apple_df.shape}')
print(f'Cleaned dataset shape: {apple_df_clean.shape}')
```

## Feature Engineering

```python
# Create lag features
apple_df_clean['lag_1'] = apple_df_clean['close'].shift(1)
apple_df_clean['lag_2'] = apple_df_clean['close'].shift(2)
apple_df_clean['lag_3'] = apple_df_clean['close'].shift(3)

# Drop rows with NaN values
apple_df_clean = apple_df_clean.dropna()

# Features and target variable
X_clean = apple_df_clean[['lag_1', 'lag_2', 'lag_3']]
y_clean = apple_df_clean['close']
```

## Split Data

```python
# Split the cleaned data into train and test sets
X_train_clean, X_test_clean, y_train_clean, y_test_clean = train_test_split(X_clean, y_clean, test_size=0.2, shuffle=False)
```

## Model Building

### Linear Regression Model

```python
# Initialize the model
lr_model_clean = LinearRegression()

# Train the model
lr_model_clean.fit(X_train_clean, y_train_clean)

# Make predictions
y_pred_lr_clean = lr_model_clean.predict(X_test_clean)

# Evaluate the model
mse_lr_clean = mean_squared_error(y_test_clean, y_pred_lr_clean)
mae_lr_clean = mean_absolute_error(y_test_clean, y_pred_lr_clean)
print(f'Cleaned Data Linear Regression MSE: {mse_lr_clean}')
print(f'Cleaned Data Linear Regression MAE: {mae_lr_clean}')
```

### Random Forest Regressor

```python
# Initialize the model
rf_model_clean = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
rf_model_clean.fit(X_train_clean, y_train_clean)

# Make predictions
y_pred_rf_clean = rf_model_clean.predict(X_test_clean)

# Evaluate the model
mse_rf_clean = mean_squared_error(y_test_clean, y_pred_rf_clean)
mae_rf_clean = mean_absolute_error(y_test_clean, y_pred_rf_clean)
print(f'Cleaned Data Random Forest MSE: {mse_rf_clean}')
print(f'Cleaned Data Random Forest MAE: {mae_rf_clean}')
```

## Visualization Of Predictions

```python
plt.figure(figsize=(12, 6))

# Plot actual vs predicted stock prices
plt.plot(apple_df_clean['date'].iloc[-len(y_test_clean):], y_test_clean, label='Actual Close Price', color='b')
plt.plot(apple_df_clean['date'].iloc[-len(y_test_clean):], y_pred_rf_clean, label='Predicted Close Price (RF)', color='r', linestyle='--')
plt.plot(apple_df_clean['date'].iloc[-len(y_test_clean):], y_pred_lr_clean, label='Predicted Close Price (LR)', color='g', linestyle='--')

plt.title('Stock Price Prediction - Actual vs Predicted')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
